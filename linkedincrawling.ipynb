```python
def language_crawling(cand_id,url):
    #Constant:
    h = HTMLParser()
    based_link = "https://media.licdn.com/mpr/mpr/shrinknp_400_400"
    language_pattern = 'urn:li:fs_language:('
    
    #initialization
    language_dict = dict()
    language_dict['cand_id'] = cand_id
    A = client.get(url)
    soup = BeautifulSoup(A.text, 'html.parser')
    language_list = list()
    for code in soup.find_all('code'):
        code_str = code.get_text()
        code_str = h.unescape(code_str)
        if 'projectView' in code_str:
            try:
                json_data = json.loads(code_str)
                profile_id = json_data['data']['profile']
                profile_id =profile_id[len('urn:li:fs_profile:'):]
                language_id = language_pattern+profile_id
                dataset = json_data['included']
                for data in dataset:
                    if 'entityUrn' in data and data['entityUrn'].find(language_id) != -1:
                        print('language name',data['name'])
                        language_dict['name'] = data['name']
                        if 'proficiency' in data:
                            language_dict['proficiency'] = data['proficiency']
                            print('language proficiency',data['proficiency'])
                        language_list.append(language_dict)
            except:
                print("error")
                            
    return language_list           
# logfile.close()
```

```python
logfile = open('Log.txt', 'w')
profilepic_string = ""
for code in soup.find_all('img'):
    code_str = code.get_text()
    code_str = h.unescape(code_str)
    code_str_list = code_str.split("\n\n\n")
    for string in code_str_list:
        #print("=================")
        string = string.replace('\n', '')
        string.strip()
        if len(string) > 3 and string[2] == '{' and string[len(string)-1] == '}':
            try:
                json_data = json.loads(string)
                #pprint.pprint(json_data)
                #jsonfile.write(json_data)
                if 'included' in json_data:
                    #print("these is included\n")
                    result_temps = json_data['included']
                    #pprint.pprint(result_temps)
                    for resutl_temp in result_temps:
                        #print("=================")
                        #pprint.pprint(resutl_temp)
                        if 'croppedImage' in resutl_temp:
                            #print(resutl_temp['croppedImage'])
                            profilepic_string = resutl_temp['croppedImage']
            except:
                logfile.write("error at:%s \n" % repr(string))
                
print(based_link+profilepic_string)
url = based_link+profilepic_string
urllib.request.urlretrieve(url, "profile_img.jpg")
logfile.close() 
```
