```python
import pandas as pd
import pprint
import CrawlingFunction as cf
```

```python
cand_df = pd.DataFrame({'cand_id': [], 'firstname': [], 'lastname': [], 'img_path': [],'url': []})
university_df = pd.DataFrame({'cand_id': [], 'name': [], 'country': [], 'startMonth': [],'startYear': [], 'endMonth': [], 'endYear': []})
```

```python
sample_df = pd.read_csv("candidate_infor_data.csv")
sample_series = sample_df.to_dict(orient = 'records')
```

```python
i = 0
nb_of_disconnect = 0
for sample in sample_series:
    print(sample['cand_id'])
    try:
        result = cf.get_user_information(sample['cand_id'], sample['url'])
        cand_row = cf.cand_info_2DF(result)
        cand_df = pd.concat([cand_df,cand_row])

        university_rows = cf.university_list_2DF(result)
        university_df = pd.concat([university_df,university_rows])
        i = i + 1
        if i % 100 == 0:
            cand_df.to_csv("cand_df_3_%d.csv"%i)
            university_df.to_csv("university_df_3_%d.csv"%i)
    except:
        nb_of_disconnect = nb_of_disconnect + 1
        if nb_of_disconnect == 10:
            break
        continue
    

```

```python
cand_df.to_csv("cand_df_3.csv")
university_df.to_csv("university_df_3.csv")
```
