```python
def language_crawling(cand_id,url):

    HOMEPAGE_URL = 'https://www.linkedin.com'
    LOGIN_URL = 'https://www.linkedin.com/uas/login-submit'

    client = requests.Session()
    h = HTMLParser()
    html = client.get(HOMEPAGE_URL).content
    soup = BeautifulSoup(html)
    csrf = soup.find(id="loginCsrfParam-login")['value']

    login_information = {
       'session_key':'nguyenvanquang247@gmail.com',
       'session_password':'conchuot123@@',
       'loginCsrfParam': csrf,
    }
    
    client.post(LOGIN_URL, data=login_information)
    
    #Constant:
    based_link = "https://media.licdn.com/mpr/mpr/shrinknp_400_400"
    language_pattern = 'urn:li:fs_language:('
    
    #initialization
    language_dict = dict()
    language_dict['cand_id'] = cand_id
    A = client.get(url)
    soup = BeautifulSoup(A.text, 'html.parser')
    language_list = list()
#     print(soup)
    for code in soup.find_all('code'):
        code_str = code.get_text()
        code_str = h.unescape(code_str)
#         print(code_str)
        if 'projectView' in code_str:
            json_data = json.loads(code_str)
            profile_id = json_data['data']['profile']
            profile_id =profile_id[len('urn:li:fs_profile:'):]
            language_id = language_pattern+profile_id
            dataset = json_data['included']
            for data in dataset:
                if 'entityUrn' in data and data['entityUrn'].find(language_id) != -1:
                    print('language name',data['name'])
                    language_dict['name'] = data['name']
                    if 'proficiency' in data:
                        language_dict['proficiency'] = data['proficiency']
                        print('language proficiency',data['proficiency'])
                    language_list.append(language_dict)

                            
    return language_list  
```

```python
def continent2ethnicity(continentcode):
    ethnicity_continent = {'EU':'White/Caucasian', 'SA':'Hispanic/Latino', 'AS':'Asian', 'AF':'Black/African American', 'NA':'White/Caucasian', 'AN':'Other/Multiracial', 'OC':'Other/Multiracial','UNKNOWN':'UNKNOWN'}
    return ethnicity_continent[continentcode]
```

```python
def languageCode2ethnicity(languagecode):
    file_path = os.path.join(os.getcwd() + '/dictionary/language_continent_dict.csv')
    print(file_path)
    language_df = pd.read_csv(file_path)
    if languagecode == 'es':
        return 'Hispanic/Latino'
    else:
        continentcode_df = language_df.loc[language_df.name == languagecode]
        if continentcode_df.empty:
            continentcode = 'UNKNOWN'
        else:
            continentcode = continentcode_df['dorminant_continent'].values[0]
    print(continentcode)
    ethnicity_continent = {'EU':'White/Caucasian', 'SA':'Hispanic/Latino', 'AS':'Asian', 'AF':'Black/African American', 'NA':'White/Caucasian', 'AN':'Other/Multiracial', 'OC':'Other/Multiracial','UNKNOWN':'UNKNOWN'}
    ethnicity_continent_notNA = {'EU':'White/Caucasian', 'SA':'Hispanic/Latino', 'AS':'Asian', 'AF':'Black/African American', 'NA':'Other/Multiracial', 'AN':'Other/Multiracial', 'OC':'Other/Multiracial','UNKNOWN':'UNKNOWN'}
    return ethnicity_continent_notNA[continentcode]
languageCode2ethnicity('pt')
```

```python
def lang_from_Country(key):
    list_lang = babel.languages.get_official_languages(key)
    list_lang_new =list()
    for lang in list_lang:
        lang_dict = dict()

        if '_'in lang:
            lang = lang.split('_')[0]
        lang_dict['code'] = lang
        try:
            language_name = pycountry.languages.get(alpha_2=lang)
            lang_dict['name'] = language_name.name
#             list_lang_new.append(language_name.name)
        except:
#             list_lang_new.append('UNKNOWN')
            lang_dict['name'] = 'UNKNOW'
        list_lang_new.append(lang_dict)

    return list_lang_new
lang_from_Country('KM')
```

```python
def langCode_from_Country(key):
    list_lang = babel.languages.get_official_languages(key)
    list_lang_new =list()
    for lang in list_lang:
        if '_'in lang:
            lang = lang.split('_')[0]
        list_lang_new.append(lang)
    return list_lang_new
langCode_from_Country('KM')
```

```python
def language2continent(lang):
    list_lang = list()
    lang = lang.lower()
    json_dir = os.path.join(os.getcwd()+'/json/country_language.json')
    # with open(json_dir,encoding='utf-8') as data_file: 
    new_lang =lang
    with open(json_dir) as data_file: 
        data = json.load(data_file)
    for language in data:
        if language['language'].lower() == lang:
            print(language['country'])
            country  = convertCountryName(language['country'])
            continent = Country2Continental(country) 
            list_lang.append(continent)
            print(continent)
    return list_lang
language2continent('vietnamese')
```

```python
def convertCountryName(name):
    shortname = ''
    json_dir = os.path.join(os.getcwd()+'/json/names.json')
    with open(json_dir) as data_file: 
        data = json.load(data_file)
    for key in data.keys():
        if data[key] == name:
            print(key)
            shortname = key
    return shortname
convertCountryName('Vietnam')
```

```python
def Country2Continental(country):
    shortname = 'UNKNOWN'
    json_dir = os.path.join(os.getcwd()+'/json/continent.json')
    with open(json_dir) as data_file: 
        data = json.load(data_file)
    for key in data.keys():
        if key == country:
            print(data[key])
            shortname = data[key]
    return shortname
Country2Continental('VN')
```

```python
def nativeLang2LangCode(lang):
    lang = lang.lower()
    json_dir = os.path.join(os.getcwd()+'/json/nativelanguage.json')
    # with open(json_dir,encoding='utf-8') as data_file: 
    new_lang =lang
    with open(json_dir) as data_file: 
        data = json.load(data_file)
        for dic in data:
            if (data[dic]['name'].lower() == lang) or (data[dic]['nativeName'].lower() == lang):
#                 print(data[dic]['name'])
                new_lang =  dic
    return new_lang
nativeLang2LangCode('Hebrew')
```
