```python
country_list = list()
for country in data['supplementalData']['territoryInfo']['territory']:
    if 'languagePopulation' in country:
        for lang in country['languagePopulation']:
            if type(lang) != str:
                if '@officialStatus' in lang:
                    country_infor = dict()
                    country_infor['language_official'] = lang['@officialStatus']
                    country_infor['country_code'] = country['@type']
                    country_infor['country_population'] = country['@population']
                    country_infor['language'] = lang['@type']
                    country_infor['language_populationPercent'] = lang['@populationPercent']
                    country_infor['language_population'] = float(country_infor['country_population'])*float(country_infor['language_populationPercent'])/100000 
                    country_infor['continent'] = Country2Continental(country_infor['country_code'])
                    country_list.append(country_infor)
```

```python
def Country2Continental(country):
    shortname = 'UNKNOWN'
    json_dir = os.path.join(os.getcwd()+'/json/continent.json')
    with open(json_dir) as data_file: 
        data = json.load(data_file)
    for key in data.keys():
        if key == country:
#           print(data[key])
            shortname = data[key]
    return shortname
Country2Continental('VN')
```

```python
def LangCode2LangName(langCode):
    langCode = langCode.lower()
    json_dir = os.path.join(os.getcwd()+'/json/nativelanguage.json')
    # with open(json_dir,encoding='utf-8') as data_file: 
    Lang_name = 'UNKNOWN'
    with open(json_dir) as data_file: 
        data = json.load(data_file)
        for dic in data:
            if dic.lower() == langCode:
                Lang_name = data[dic]['name']
    return Lang_name
LangCode2LangName('es')
```
