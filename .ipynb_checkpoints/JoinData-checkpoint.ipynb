```python
import pandas as pd
import pprint
```

```python
sample_df = pd.read_csv('sample_data_1000_8.csv')
sample_df =sample_df.rename(columns={'cand_id':'globalID'})
sample_df = sample_df.rename(columns={'ID':'cand_id'})
sample_df.head()
```

```python
com_df = pd.read_csv('com_df.csv')
all_df = pd.merge(com_df,sample_df,on='cand_id',how='left')
compare_df = all_df[['cand_id', 'country','ethnicity','candidate_country']]
compare_df
compare_df.to_csv('compare_df_result.csv')
```

```python
uni_df = pd.read_csv('university_df.csv')
all_df = pd.merge(uni_df,sample_df,on='cand_id',how='left')
compare_uni_df = all_df[['cand_id', 'country','ethnicity','candidate_country']]
compare_uni_df
compare_uni_df.to_csv('compare_uni_df_result.csv')
```

```python
lang_df = pd.read_csv('language_df.csv')
all_df = pd.merge(lang_df,sample_df,on='cand_id',how='left')
compare_lang_df = all_df[['cand_id', 'name','proficiency','ethnicity','candidate_country']]
lang_Native_df = compare_lang_df.loc[compare_lang_df.proficiency == 'NATIVE_OR_BILINGUAL']
lang_Native_df
```
