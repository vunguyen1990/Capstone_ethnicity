```python
based_url = "https://www.ethnologue.com"
general_map_url = '/browse/maps'
r  = requests.get(based_url + general_map_url)
data = r.text

soup = BeautifulSoup(data)
```

```python
def subcountryFromLang(ref_url):
    r  = requests.get(based_url + ref_url)
    data = r.text
    soup = BeautifulSoup(data)
    for subdata in soup.findAll('div', { "class" : "views-field views-field-field-national-languages" }):
        if 'Principal Languages' in subdata.text:
            for div in subdata.findAll('div', { "class" : "field-content" }):
                list_language = div.text.split(',')
            break
    return list_language
list_lang = subcountryFromLang('/country/DZ')
count = 0
for lang in list_lang :
    print("%s, %d"%(lang, count))
    count = count +1
```

```python
def countryFromLang(ref_url):
    list_lang = []
    r  = requests.get(based_url + ref_url)
    data = r.text
    soup = BeautifulSoup(data)
    for subdata in soup.findAll('div', { "class" : "field-item even" }):
       
        if 'See language listings for' in subdata.text:
            print('========================\n')
            print(subdata.text)
            print('========================\n')
            for link in subdata.find_all('a'):
                print(link.get('href'))
                result =subcountryFromLang(link.get('href'))
                list_lang = list_lang + result
#                 joinlist = list(set(list_lang + link.get('href')))
#                 list_lang.append(subcountryFromLang(link.get('href')))
            break
    all_list = list_lang
    return all_list

all_list = countryFromLang('/map/DZMATN')
```
